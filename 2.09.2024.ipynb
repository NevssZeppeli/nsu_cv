```python
A =  np.zeros(img.shape, dtype=np.uint8)


for r in range(int(A.shape[0]/2-5),int(A.shape[0]/2+6)):
    #print(abs(r-A.shape[0]/2)/10)
    #print((str( int(-5 *np.cos(np.pi * abs(r-A.shape[0]/2)/10))) + " ; " + str(int(5*np.cos(np.pi * abs(r-A.shape[0]/2)/10)))))
    for i in range(int(-5*np.cos(np.pi * abs(r-A.shape[0]/2)/10)) , int(5*np.cos(np.pi * abs(r-A.shape[0]/2)/10))):
        A[r,int(i+A.shape[1]/2)] = 255
        #print(str(r)+ " ; " + str(i))

plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(A, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img = cv.imread(cv.samples.findFile("jaba.jpg"), cv.IMREAD_GRAYSCALE)
```

```python
dft = cv.dft(np.float32(img),flags = cv.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)
 
magnitude_spectrum = np.log(np.log(cv.magnitude(dft_shift[:,:,0],dft_shift[:,:,1])))
```

```python
FA = cv.dft(np.float32(A),flags = cv.DFT_COMPLEX_OUTPUT)
M = np.log(cv.magnitude(FA[:,:,0],FA[:,:,1]))
MS = np.fft.fftshift(M)


MS*magnitude_spectrum



plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow((MS*magnitude_spectrum), cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
im_eg = cv.idft(np.float32(MS*magnitude_spectrum),flags = cv.DFT_COMPLEX_OUTPUT)

plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(im_eg, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```
